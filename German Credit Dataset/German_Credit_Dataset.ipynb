### Loan Approval Prediction

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
import pydotplus as pdot
from IPython.display import Image
```

## Importing Classificaions Models 

```python
# KNN classification Model 
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()

# Gaussian Navi Bayes Model
from sklearn.naive_bayes import GaussianNB
nbc = GaussianNB()

## SVM Model 
from sklearn import svm
svmc = svm.SVC(kernel = "linear")

# Decision Tree Classifier 
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state = 42)

# Random Forest 
from sklearn.ensemble import RandomForestClassifier
forest_clf = RandomForestClassifier()

# Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Linear SVC
from sklearn.svm import LinearSVC
linsvc_clf = LinearSVC()

# Bagging Classifier
from sklearn.ensemble import BaggingClassifier
bgcl = BaggingClassifier(n_estimators = 100, max_samples = 100,bootstrap = True)
```

## Importing Efficiency measurement metrics

```python
credit_df = pd.read_csv( "German Credit Data.csv" )
credit_df.info() 
```

```python
x_features = list( credit_df.columns )
x_features.remove( 'status' )
x_features 
```

```python
encoded_credit_df = pd.get_dummies( credit_df[x_features],drop_first = True)
```

```python
encoded_credit_df.columns
```

```python
y = credit_df.status
x = encoded_credit_df
```

```python
x.head()    
```

## Splitting data for training and testing 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify = y, random_state = 1, test_size = 0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train.head()
```

**bold text**## Training and testing using different Classification Models 

### Ensembling

```python
for model in (knn, nbc, svmc, dtc, forest_clf, linsvc_clf):
    model.fit(x_train, y_train)
    y_pred_train = model.predict(x_train)
    y_pred_test = model.predict(x_test)
    print("The model is :", model)
    print("Training Accuracy is : ", accuracy_score(y_train,y_pred_train))
    print("Test Accuracy is : ", accuracy_score(y_test,y_pred_test))
    print("Training ROC_AUC score is:", roc_auc_score(y_train, y_pred_train))
    print("Test ROC_AUC score is:", roc_auc_score(y_test, y_pred_test))
    print("-----------------------------------------------------------")
```

### Decision Tree Classifier is over fitting 



```python
dtc = DecisionTreeClassifier(random_state = 42, max_depth = 5)
forest_clf = RandomForestClassifier()

for model in (knn, nbc, svmc, dtc, forest_clf, linsvc_clf):
    model.fit(x_train, y_train)
    y_pred_train = model.predict(x_train)
    y_pred_test = model.predict(x_test)
    print("The model is :", model)
    tra_acc = accuracy_score(y_train,y_pred_train)
    tes_acc = accuracy_score(y_test,y_pred_test)
    tra_roc_auc = roc_auc_score(y_train, y_pred_train)
    tes_roc_auc = roc_auc_score(y_test, y_pred_test)
    print("Training Accuracy is : ", tra_acc)
    print("Test Accuracy is : ", tes_acc)
    print("Training ROC_AUC score is:", tra_roc_auc)
    print("Test ROC_AUC score is:", tes_roc_auc)
    print("-----------------------------------------------------------")
```
